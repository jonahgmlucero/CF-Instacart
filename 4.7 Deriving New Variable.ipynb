```python
#creating a subset -- the first one million rows
df=ords_prods_merge[:1000000]
```

```python
#conditioning the data
def price_label(row):

    if row['prices']<= 5:
        return 'Low-range product'
    elif (row['prices']>5) and (row['prices'] <15):
        return 'Mid-range product'
    elif row['prices'] >15:
        return 'High range'
    else: return 'Not enough data'
```

```python
#using the price_label function
df['price_range']=df.apply(price_label, axis=1)
```

```python
#for-loop in the Instacart dataframe. Create a new column in the 
#ords_prods_merge dataframe to show how busy each day of the week is
```

```python
#1. print the frequency of the 'orders_day_of_week' column
ords_prods_merge['orders_day_of_week'].value_counts(dropna=False)
```

```python
#0=saturday, busiest day; 4=Wednesday, slowest day
```

```python
#creating a new column "busiest day"
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#print the result
result
```

```python
#now merge the new column with the ords_prods_merge dataframe
ords_prods_merge['busiest_day']=result
```

```python
ords_prods_merge['busiest_day'].value_counts(dropna=False)
```

# Task Portion

```python
#supposed client would like to label the two busiest days as "busiest days" and two slowest days as "slowest days"
#according to orders_days_of_week; Busiest days: Saturday(0) and Sunday(1)
#Slowest days: Tuesday(3) and Wednesday(4)
#They're asking for a new column for this

result2 = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result2.append("Busiest days")
  elif value == 3 or value == 4:
    result2.append("Slowest days")
  else:
    result2.append("Regularly days")
```

```python
#print result2
result2
```

```python
#now merge the new column with the ords_prods_merge dataframe
ords_prods_merge['busiests_days'] = result2
```

```python
ords_prods_merge['busiests_days'].value_counts(dropna=False)
```

#### Identifiying the busiest hour of the day. Label the time period as "Most Orders", "Average Orders", and "Fewest Orders"
#### Label the column as "busiest_period_of_day"

```python
#for-loop has to be done using order_timestamp column
ords_prods_merge['order_timestamp'].value_counts(dropna=False)
```

```python
#order_timestamp is divided to 3 and grouped according to how busy it is from most busy to least busy
result3 = []

for value in ords_prods_merge["order_timestamp"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        result3.append("Most Orders")
    elif value in [17, 8, 18, 19, 20, 7, 21, 22]:
        result3.append("Average Orders")
    else:
        result3.append("Fewest Orders")
```

```python
#print result3
result3
```

```python
#now merge the new column with the ords_prods_merge dataframe
ords_prods_merge['busiest_period_of_day'] = result3
```

```python
ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
